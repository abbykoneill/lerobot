## Generative AI with ChatGPT
ChatGPT is a well-known generative AI chatbot language model tool used for everyday use including writing, planning, answering questions, explaining instructions, reading images, and more. We introduce it here for a high level understanding of the capabilities of Gen AI tools.

For this part of the lab, you will explore basic functionality with ChatGPT to gain an understanding of how powerful these language models can be, even in everyday use cases.

Open [ChatGPT](https://chatgpt.com/) and follow the prompts in this notebook. Make sure to save your outputs and discuss answers to the questions with your lab partner.

## (A) Content Explanation

1. Pick any concept you remember from a previous course and ask ChatGPT to explain it to you. Record what you asked and how ChatGPT responded. How well does the answer match how you would have explained this concept to a peer?

>**Note:** Your prompt should follow a similar structure to: _Explain..._

Below are a few topic ideas, or create your own.
* _Explain..._ photosynthesis
* _Explain..._ compiling code
* _Explain..._ the Oedipus Complex

Notice how ChatGPT will conclude it's answer by asking about continuing the conversation. "_If you'd like, I can also:_"
2. Choose one of the presented options for your concept and ask ChatGPT to explain further. Record what you asked and how ChatGPT responded.
